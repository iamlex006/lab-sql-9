```python
#Q1 How many distinct (different) actors' last names are there?
```

```python
pd.set_option('display.max_rows', None)
```

```python
actor = pd.read_sql_query('''SELECT last_name, COUNT(last_name)
FROM sakila.actor
GROUP BY (last_name)
HAVING COUNT(last_name) = '1';''', engine)
actor
```

```python
#Q2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
pd.set_option('display.max_rows', 30)
```

```python
day = pd.read_sql_query('''SELECT rental_id, rental_date, inventory_id, customer_id, return_date, staff_id, last_update, MONTH(rental_date) AS month, weekday(rental_date) AS weekday,
	CASE
	WHEN ((weekday(rental_date)) >=0) and ((weekday(rental_date))< 5) then 'workday'
	WHEN ((weekday(rental_date)) >4) and ((weekday(rental_date))<= 6) then 'weekend'
	ELSE 'Error'
	END AS 'day_type'
FROM sakila.rental;;''', engine)
day
```

```python
#Q3 Get all films with ARMAGEDDON in the title.
```

```python
arma = pd.read_sql_query('''SELECT *
FROM sakila.film
WHERE title LIKE '%%ARMAGEDDON%%';''', engine)
arma
```

```python
#Q4 Get 10 the longest films
```

```python
ten = pd.read_sql_query('''SELECT *
FROM sakila.film
ORDER BY length DESC
LIMIT 10;''', engine)
ten
```

```python
#Q5 How many films include Behind the Scenes content?
```

```python
bts = pd.read_sql_query('''SELECT COUNT(*)
FROM sakila.film
WHERE special_features LIKE '%%Behind The Scenes%%';''', engine)
bts
```

```python
#Q6 Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
length = pd.read_sql_query('''SELECT rating, ROUND(AVG(length),2) AS mean
FROM sakila.film
GROUP BY (rating)
HAVING mean > 120
ORDER BY (rating);''', engine)
length
```

```python
# Q7 Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
```

```python
rank = pd.read_sql_query('''SELECT title, length, RANK() OVER(ORDER BY (length) DESC) AS ranking
FROM sakila.film;''', engine)
rank
```
